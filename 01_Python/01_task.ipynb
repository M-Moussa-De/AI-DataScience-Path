```python
import math


# Validate user's numeric input
def read_input(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Invalid input. Please enter a valid number.")
```

## **1- Write a Python program to find the area of a triangle**

```python
# Read triangle sides
a = read_input("Enter side a: ")
b = read_input("Enter side b: ")
c = read_input("Enter side c: ")

# Check if valid triangle
if a + b > c and a + c > b and b + c > a:
    s = (a + b + c) / 2  # semi-perimeter
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))
    print("Area of the triangle is:", area)
else:
    print("Invalid triangle sides")
```

# **2- Write a Python program to swap two variables.**


```python
def swap(num1, num2):
   temp = num1
   num1 = num2
   num2 = temp
   return num1, num2

a = read_input("Enter 1st number: ")
b = read_input("Enter 2nd number: ")

swap(a, b)
```

# **3-Write a Python program to enter a number from user and convert this number from kilometers to miles.**


```python
km = read_input("Enter KM number: ")
miles = km * 0.621371
print(f"{km} km is equavalent to {miles} miles")
```
